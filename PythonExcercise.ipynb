```python
#######

# Python While Loops
# while loops run a block of code as long as a condition is true.
# Be careful with while loops—if the condition never becomes False, it creates an infinite loop.
# Infinite loops can freeze your program and should be avoided unless intentional.
# It's important to update the variable used in the condition (like i = i + 1) within the loop.

# Also, Python uses indentation to define blocks of code.
# Code inside the while loop must be indented consistently.
# Indenting incorrectly will either cause an error or unexpected behavior.

# Variables defined inside the while loop (inside the indentation) can be used outside if they were declared earlier.
# Example:
i = 1
while i <= 5:
    print(i)
    i = i + 1  # Make sure to update the condition variable

print("Loop finished. Final value of i:", i)

# Try It Yourself:
# Use a while loop to print numbers from 10 down to 1
# Create a variable starting from 10 and decrease it by 1 in each iteration
# Use a while loop to print numbers from 10 to 1 (countdown)

#######
```

```python
# Python Functions
# Functions let you group code into reusable blocks.
# Think of a function like a machine: you give it some inputs, it processes them, and gives you an output.
# Functions make your code reusable, easier to test, and more organized.

# Example 1: Simple greeting function
def greet(name):
    print("Hello, " + name)

greet("Sara")

# Example 2: A function that calculates area of a square
# We'll define a variable inside the function, then return it

def area_of_square(side):
    area = side * side
    return area

print("Area of square:", area_of_square(4))

# Example 3: A function that calculates area of a rectangle
# Same concept, define variables inside before returning

def area_of_rectangle(length, width):
    area = length * width
    return area

print("Area of rectangle:", area_of_rectangle(5, 3))

# Example 4: Variable scope
# Variables defined inside a function are local to that function
x = 10  # global variable

def demo():
    x = 5  # local variable
    print("Inside function, x =", x)

demo()
print("Outside function, x =", x)

# Try It Yourself:
# Create a function that takes two numbers, adds them, and returns the result
# Call the function with two values and print the result
# Create another function that returns the square of a number
# Try printing a global variable and a local variable with the same name to see the difference
# Create a function that takes two numbers, adds them, and returns the result
# Call the function with two values and print the result

#######
```

```python
#######

# Python Dates
# Python has a built-in module called `datetime` for working with dates and times.
# You can get the current date, define specific dates, add or subtract days, compare dates, and format them as strings.

import datetime

# Example 1: Getting today's date
today = datetime.date.today()
print("Today's date:", today)

# Example 2: Creating a specific date
independence_day = datetime.date(2023, 8, 14)
print("Independence Day:", independence_day)

# Example 3: Adding days to a date
from datetime import timedelta
future_date = today + timedelta(days=7)
print("7 days from today:", future_date)

# Example 4: Comparing dates
if today > independence_day:
    print("Independence Day has passed this year.")
else:
    print("Independence Day is coming up.")

# Example 5: Converting a date to string
formatted_date = today.strftime("%B %d, %Y")  # Example: May 16, 2025
print("Formatted date:", formatted_date)

# Try It Yourself:
# Create a date for your birthday using datetime.date()
# Add 30 days to it and print the new date
# Compare your birthday to today's date and print whether it has passed or is upcoming
# Convert both dates to strings and print them using strftime()

#######
```
